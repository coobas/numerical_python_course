## Note to the beginning: mutable versus immutable 
Data types can *mutable* (changeable) or *immutable* (unchangeable). Immutable objects (in Python, basically anything is an object - we will see this many times again) can not change their values. In contrast, mutable objects can change their value (without losing their identities). Immutable types are numbers (`int`, `float`, `complex` etc.), strings (`str` and `unicode`) and `tuple`. Mutable types are eg. `list`, `dict` or `set`.


# Tuple

```python
tuple1 = (1, 'a', 5)           # The basic syntax for creating a tuple (parentheses) 
tuple2 = 1, 'a'                # Parentheses are not required, but ...! 
tuple3 = tuple(["a", "b"])     # Advanced: Create a tuple from another container 
tuple4 = tuple(xrange(0, 10))  # Advanced: Create a tuple from an iterator / generator 
tuple5 = ()                    # an empty tuple 
tuple6 = ("single", )          # A tuple with one element 
tuple7 = 0, "1", (0, 1, 2)     # Tuple can of course contain another tuple

# And what we got out?
print("tuple1={}".format(tuple1))
print("tuple2={}".format(tuple2))
print("tuple3={}".format(tuple3))
print("tuple4={}".format(tuple4))
print("tuple5={}".format(tuple5))
print("tuple6={}".format(tuple6))
print("tuple7={}".format(tuple7))
```

## Unpacking
Tuple can be used to assign values to multiple variables simultaneously. For example,

```python
(x, y, z) = (1, 2, 3)
print(y)
```

In this case, the paretheses are often omitted, so we can write

```python
x, y, z = 1, 2, 3
print(y)
```

This is especially useful for functions that return multiple values:

```python
def neighbors(x):
    """Returns the integers a and b less and greater than x, i.e. a < x < b"""
    from math import ceil, floor
    a = int(floor(x))
    b = int(ceil(x))
    # 1 must be added / subtracted if x is an integer
    if a == x:
        a -= 1
        b += 1
    return a, b
# we see that the function returns a tuple
print(neighbors(-1))
x = 3.3
# now assign the result to two variables
a, b = neighbors(x)
print("%g < %g < %g" % (a, x, b))
```

Any iterable object can appear on the right hand side (more on iterators later), for example, `list` (see below) or string.

```python
# elements from a list
a, b, c = [1, 2, 3]
print(a, b, c)
# or a string
a, b, c = "123"
print(a, b, c)
```

Python 3 adds a very useful functionality of *extended unpacking*.

```python
%%python3
# c is assigned the remaining elements in the form of a list
a, b, *c = (1, 2, 3, 4, 5, 6)
print(a)
print(b)
print(c)
```

Important is of course the *asterisk*, which may be also in the middle.

```python
%%python3
a, *b, c = (1, 2, 3, 4, 5, 6)
print(a)
print(b)
print(c)
```

Python 2 throws an exception if the number of elements is incorrect.

```python
a, b, c = (1, 2, 3, 4, 5, 6)
```

# List
List is similar to tuple, however, is mutable, ie., one can change its elements. It is created using brackets [ ] or the `list` function.

```python
list(), []                 # en empty list
list1 = ["a", "b", "c"]    # list is cretaed using [...]
list2 = [0, 0.0, "0.0"]    # list can contain arbitrary types
list3 = list(tuple1)       # list can be created from a tuple

print(list1)
print(list2)
print(list3)
```

List containt more methods than tuple, which follows from the fact that it is mutable.

```python
", ".join(item for item in dir(list) if not item.startswith("_"))
```

Natural is adding to the list at the end with `append` and removing from the end with `pop`:

```python
list1.append("d")         # adding an element
print(list1)              # list1 has changed!
list1.sort(reverse=True)
print(list1)
print(list1.pop())        # pops (removes and returns) the last element
print(list1)
```

Elements can also be removed using `remove` but this method must search for the element first.

```python
list1.remove("d")         # removing element(s)
print(list1)
```

Nested lists can create "multidimensional" lists.

```python
l = [[11, 12], [21, 22]]   # "vícerozměnrý" list
print(l[0][0])             # prvek [0,0]
```

All mutable types (and therefore the list as well) are basically references or, if you prefer, pointers (pointers). Bare in mind this in order not to *accidentally overwrite the contents of another variable*!

```python
a = [1, 2, 3]
b = a            # b is identical to a (not a copy)
b.insert(0, 0)   # as list is mutable, a is modified as well
print(a)
```

```python
print(a is b)          # the is operator tests the identity (of objects)
from copy import copy
b = copy(a)            # the copy module can be used to create copies
print(a is b)
print(a == b)          # the == operator tests the values
b.append(5)
print(a)
print(b)
```

## Indexing aka *slicing*

Slicing is a very important concept. For variables of type `list` and `tuple`, slices can be used to select element(s) in a sophisticated manner; they can be used to change the list as well. `list` and `tuple` allow so called simple slice, for details see the [documentation](http://docs.python.org/2/reference/expressions.html#slicings). Eextended slicing will be used later for Numpy arrays. The syntax of a simple slice is

    [lower_bound] ":" [upper_bound] [":" [stride]]
   
The default value for the upper and lower bounds is None, the default stride value is 1. The result is an element with indices from the lower bound (inclusive) to the elements with indices smaller than the upper bound, possibly with a given step (stride). Let's see how it works in examples.

```python
# create a simple list (range in Python 3 does not return list, so it is better to use conversion)
l = list(range(10))

# all elements
print(l[:])

# the first two elements
print(l[0:2])

# the last three elements
print(l[-3:])

# even elements
print(l[::2])

# elements in reversed order
print(l[::-1])
```

With slices we can add elements to a list (there also exists the `insert` method).

```python
l = list(range(10))
a = l[:]
a[:] = []
print(l)
```

```python
l = list(range(10))   
print(l)
l[:0] = ["the first one"]
print(l)
# we can replace several elements with others, whose number may not be the same
l[1:3] = ["one", "two", "three"]
print(l)
```

... or delete elements

```python
l = list(range(10))   
print(l)
# deletes [0:2]
l[0:2] = []
print(l)

# del can be used as well
del l[0:2]
print(l)
```

```python
l = list(range(10))
%timeit l[0:2] = []
```

```python
l = list(range(10))
%timeit del l[0:2]
```

We already know that `list` is *mutable* and that a copy can be created using the `copy` module. We can equally well use the [:] slice, ie.

```python
a = ["a"]
b = a[:]
# are a and b identical?
print(a is b)
```

Dictionary (slovník)

Slovník je asociativní pole, jehož klíči jsou jakékoliv hashovatelné objekty (čísla, řetězce, tuply a většina uživatelsky definovatelných tříd). Jako klíč se nedají použít jiné kontejnery (které jsou nehashovatelné).

### Další varianty slovníku

#### collections.OrderedDict
V běžném slovníku jsou data neřazená. Pokud tedy procházíte slovníkem prvek po prvku, klíče budou dosti přeházené (podle toho, jak je zpracuje interní hashovací funkce). Alternativou je OrderedDict (seřazený slovník), který prvky za sebou řadí. Ovšem pozor: Řadí je podle toho, v jakém pořadí byly do slovníku vloženy, nikoliv podle abecedy nebo jiného kritéria.

```python
import collections
pairs = (("a" , 0), ("b", 1), ("c", 2), ("d", 3), ("e", 4))      # tuple tuplů
normal = dict(pairs)
ordered = collections.OrderedDict(pairs)

print("Keys (normal dictionary): " + ", ".join(key for key in normal))
print("Keys (ordered dictionary): " + ", ".join(key for key in ordered))
```

## Pole v Numpy

Budeme trochu předbíhat, ale univerzalita základních kontejnerů může být v mnoha případech (při vědeckých výpočtech) nevýhodou - co se týče rychlosti, ale i vzhledem k tomu, že nemáme zaručený typ prvků ani rozměry pole. Proto knihovna numpy zavádí typ vícerozměrného pole ("ndarray"), který je extrémně rychlý (vektorové operace s ním jsou implementovány v C) a velice šikovný (umožňuje snadno pracovat s mnohorozměrnými poli i jejich komplikovanými indexy). Seznámíme se s ním v některé z dalších hodin.

Na jednoduché ukázce si ukážeme, jak se liší rychlost sčítání vektorů bez a s numpy.

```python
array1 = list(range(0, 1000000))
array2 = list(range(1000000, 2000000))
print("Bez numpy")
%timeit array3 = [array1[i] + array2[i] for i in range(0, len(array1))]
```

```python
import numpy as np
array1 = np.arange(0, 1000000)
array2 = np.arange(1000000, 2000000)
print("S numpy")
%timeit array3 = array1 + array2
```

Jak vidíte, stejná operace (sečtení pole prvek po prvku) s využitím numpy proběhla asi 20x rychleji.

## Vestavěné funkce pro práci s kontejnery
Python má několik důležitých funkcí, které se hodí pro práci s kontejnery.

`len` vrací počet prvků

```python
o = [1, 1, 2, 2]
print("len(%s) = %i" % (o, len(o)))
o = 1, 1, 2, 2
print("len(%s) = %i" % (o, len(o)))
o = {1, 1, 2, 2}
print("len(%s) = %i" % (o, len(o)))
```

`sum` vrací součet prvků

```python
o = [1, 1, 2, 2]
print("sum(%s) = %i" % (o, sum(o)))
o = 1, 1, 2, 2
print("sum(%s) = %i" % (o, sum(o)))
o = {1, 1, 2, 2}
print("sum(%s) = %i" % (o, sum(o)))
```

Pozor, python je silně typovaný jazyk

```python
o = 1, 1, 2, 2, "3"
print("sum(%s) = %i" % (o, sum(o)))
```

`min` a `max` vrací nejmenší, resp. největší prvky.

```python
o = [1, 2, -1, -10, 0]
print("min(%s) = %i" % (o, min(o)))
print("max(%s) = %i" % (o, max(o)))
```

`sorted` vrací setříděné prvky pole.

`reversed` vrací prvky pozpátku (pomocí iterátoru).

### `all`, `any` a meší odbočka k `bool`
`all` a `any` vrátí logické `and`, resp. `or` aplikované mezi všemi prvky.

```python
o = [True, True, True]
print("all(%s) = %s" % (o, all(o)))
print("any(%s) = %s" % (o, any(o)))
o = [True, False, True]
print("all(%s) = %s" % (o, all(o)))
print("any(%s) = %s" % (o, any(o)))
```

Při této příležitosti ještě odbočíme a ukážeme si, jak Python převádí cokoli (tj. objekt jakéhokoli typy) na `bool`, tj. na `True` nebo `False`. Tento převod můžeme udělat explicitně pomocí samotné funkce `bool`, děje se ale také implicitně v blocích typu `if` nebo `while` a také při použití `all` a `any`.

všechna čísla kromě 0 jsou True

```python
print(bool(0))
print(bool(0.0))
print(bool(0.0 + 0j))
print(bool(-1))
```

řetězce jsou True pokud nejsou prázdné

```python
print(bool(""))
print(bool("0"))
```

kontejnery `tuple`, `list`, `dict`, `set` apod. jsou True pokud nejsou prázné

```python
print(bool([]))
print(bool(()))
print(bool({False}))
# pokud chceme otestovat jednotlivé prvky, musíme použít all nebo any
print(all({0}))
```

toto už možná není zcela intuitivní
