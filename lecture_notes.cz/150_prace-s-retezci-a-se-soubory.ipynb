### f-strings

Python 3.6 zavedl moderní způsob formátování řetězců pomoci tzv. [f-strings](https://docs.python.org/3/reference/lexical_analysis.html#f-strings). Syntaxe je velice jednoduchá: před první uvozovky vložit znak `f` a pak všechny výrazy uvnitř `{složených závorek}` nahradit jejich textovou reprezentací.

Např.:

```python
name = "Eric"
age = 74
f"Hello, {name}. You are {age}."
```

```python
f"2 * 37 = {2 * 37}"
```

Můžeme volat i metody / funce:

```python
name = "Eric Idle"
f"{name.lower()} is funny."
```

Výrazy se dají formátovat pomocí [formátovacího mini-jazyka](https://docs.python.org/3/library/string.html#formatspec):

```python
import decimal

value = decimal.Decimal("12.34567")
width = 10
precision = 4

f"result: {value:{width}.{precision}}"
```

```python
import datetime as dt

today = dt.date.today()
f"Taday is {today:%B %d, %Y}"
```

Existují ještě dva starší (zastaralé) způsoby formátování: pomocí `%` operátoru a pomocí `format` metody. Doporučení je používat už jen f-strings, případně [`string.Template`](https://docs.python.org/3/library/string.html#template-strings). Viz např. https://realpython.com/python-string-formatting/#which-string-formatting-method-should-you-use.

### Metody třídy `str`
Výpis "veřejných" metod `str`, k němuž použijeme `join` pro vytvoření řetězce z iterátoru a `startswith` pro test začátku řetězce.

```python
", ".join([m for m in dir("") if not m.startswith("_")])
```

Mezery apod. ořízneme pomocí `strip`.

```python
s = "  \t abc\n\n"
print(s)
print(s.strip())
```

Často se hodí split, je to vlastně opak join.

```python
s = "1, 2, 3, čtyři, 5"
# rozdílema na částí oddělené čárkou
s = s.split(",")
print(s)
# zkusíme převést na čísla
s = [int(a) for a in s if a.strip().isdigit()]
print(s)
```

Velikost písmen mění `upper`, `lower`, `capitalize`, `swapcase`

```python
print("všechna písmena velká".upper())
```

## Základy práce se soubory
Pro práci se sounory je základní příkaz `open`, který vrací instanci `file`

```python
f = open("./150_prace-s-retezci-a-se-soubory.ipynb", "r")
print(f)
```

```python
# metody file
", ".join([m for m in dir(f) if not m.startswith("_")])
```

```python
# teď soubor zavřeme
f.close()
```

Nejčastější je procházet soubor jako iterátor.

```python
f = open("./150_prace-s-retezci-a-se-soubory.ipynb", "r")

# print first 10 lines with line numbers and max 20 characters per line
for count, line in enumerate(f):
    print(f"{count}: {line[:20].rstrip()}")
    if count == 10:
        break
f.close()
```

Pro práci se soubory je velice užitečný (*a jednoznačně preferevaný*) blok [`with`](https://docs.python.org/3/reference/compound_stmts.html#with), který používá tzv. context managery. To nám zajistí, že soubor bude vždy zavřen, i když během práce se souborem dojde k chybě (neodchycená výjimka). Nahrazuje to tak `try..except..finally` blok. 

Použití je velice jednoduché, předchozí příklad by vypadal takto:

```python
with open("./150_prace-s-retezci-a-se-soubory.ipynb", "r") as f:
    # print first 10 lines with line numbers and max 20 characters per line
    for count, line in enumerate(f):
        print(f"{count}: {line[:20].rstrip()}")
        if count == 10:
            break
```

Všimněte si, že chybí explicitní volání `f.close()`.
