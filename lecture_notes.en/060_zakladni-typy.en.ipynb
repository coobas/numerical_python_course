# Numeric types
Python has built-in four basic numeric types: 

1. integers: `int`, `long` (Python 3 combines them into int) 
2. decimal numbers: `float` (equivalent to double in C) 
3. complex numbers: `complex`
4. logical (boolean): `bool`

```python
1 + 2 - 3     # integers
```

```python
int1 = 10     # decimal notation
int2 = 0b10   # binary
int3 = 0x10   # hexadecimal
int4 = 0o10   # octal
print("%i, %i, %i, %i" % (int1, int2, int3, int4))
print("The same binary")
print("%s, %s, %s, %s" % (bin(int1), bin(int2), bin(int3), bin(int4)))
```

```python
print(4 / 3)  # floating point division (always in Python 3)
print(4 // 3)  # the integer division operator
print(4.0 // 3.0)  # // works for floats as well
```

```python
print("type(1) = %s" % type(1))
# Long is for big numbers
print("type(10000000000000000000) = %s" % type(10000000000000000000))   
# Big numbers can be realllly big
big_number = 999**999    
print("999^999 = %d" % big_number)
print("999 ^ 999 has {} digits.".format(len(str(big_number))))
```

In Pythonu 2, / is integer division for integer

Python 2 can act as a Python 3 using the `__future__` module.

Calculations involving floating point numbers are as usual, watch the final accuracy.

```python
3.1 * (0.2 - 0.1)**2
```

Python knows complex numbers.

```python
abs(1 - 1j)
```

The Boolean values are easily handled.

```python
b = True      # boolean
print(not b)  # logical operator not
```

# Strings -- `string` and `unicode`
Python includes  `str` and `unicode` types (Python 3 unifies to `builtins.str`) and a rich set of features for working with them. Those are implemented either as methods (more about classes and methods later) or as functions in modules `string`, `StringIO`, `re` and others.

```python
a = 3 * ("'Abc'%i" % 3) +'!'  # a possible way of creating a string
print(a)
```

One can access specific parts of a string using indexes.

```python
print(a[0])        # the first charactes
print(a[0:-1])     # all characters except the last one
print(a[-2:])  # the last character (None is optinal here)
```

The [`format`](https://docs.python.org/2/library/stdtypes.html#str.format) method enables advanced text formatting.

```python
a = '{0}{1}{0}'.format('abra', 'cad')  # the format method
print(a)
print(a.upper())  # upper case
print(a.find("ra"))  # simple search
print(a.upper().split('A'))  # splitting be a specific character
```

The [`string`](https://docs.python.org/2/library/string.html) module contains more funstions.

```python
import string       # import the module
print(dir(string))  # print out its contents
```

Let's try `capwords`
